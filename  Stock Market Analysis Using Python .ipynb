```python
#Add a new column ‘Trend’ whose values are based on the 'Daily Return
#This is a function that takes a numerical value as input and returns a string that describes the trend of the value
def trend(x):
    if -0.5 <= x <= 0.5:
        return 'Slight No change'
    elif 0.5 < x <= 1:
        return 'Slight Positive'
    elif -1 <= x < -0.5:
        return 'Slight Negative'
    elif 1 < x <= 3:
        return 'Positive'
    elif -3 <= x < -1:
        return 'Negative'
    elif 3 < x <= 7:
        return 'Among top gainers'
    elif -7 <= x < -3:
        return 'Among top losers'
    elif x > 7:
        return 'Bull run'
    else:
        return 'Bear drop'
```

```python
df['Apple_Trend'] = daily_return_Apple.apply(lambda x:trend(x))
df['Microsoft_Trend'] = daily_return_Microsft.apply(lambda x :trend(x))
df['Google_Trend']= daily_return_Google.apply(lambda x :trend(x))
df['Amazon_Trend'] = daily_return_Apple.apply(lambda x :trend(x))
```

```python
#What was the correlation between the daily returns of different stocks
daily_return_avg.dropna(inplace= True, axis = 0)
daily_return_avg.iloc[1: , :]
```

```python
corr_df = daily_return_avg.corr(method='pearson')
corr_df.head().reset_index()
corr_df.head()
```

```python
from matplotlib.colors import ListedColormap

# Define PESTLE colors
pestle_colors = ['firebrick', 'orange', 'gold', 'seagreen', 'royalblue']

# Define custom color map
cmap = ListedColormap(pestle_colors)

# Compute correlation matrix and create mask
corr_df = daily_return_avg.corr(method='pearson')
mask = np.zeros_like(corr_df)
mask[np.tril_indices_from(mask)] = True

# Plot heatmap with custom color map
sns.heatmap(corr_df, cmap=cmap, vmin=-1.0, vmax=1.0, mask=mask, linewidths=2.5)
plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.show()
```
